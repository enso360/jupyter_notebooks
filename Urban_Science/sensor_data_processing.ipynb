```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = "data/node_99.csv"
start_time_string = '2017-03-20 12:15:00'
end_time_string = '2017-03-20 19:25:00'

df = pd.read_csv(filename)
df
```

```python
# add average temperature column for t1 and t2
df['t_avg'] = df[['t1', 't2']].mean(axis=1)
df.head()
```

```python
# create a new df for data resampling
data = pd.concat([df['timestamp'], df['t_avg']], axis=1, keys=['timestamp', 't_avg'])

# set the index to DateTime object
data = data.set_index('timestamp')
data.index = pd.to_datetime(data.index)
data.index
data

# slice the data as per time interval required to match index of LSI meter data
data = data.loc[start_time_string : end_time_string]
data
```

```python
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.core.resample.Resampler.fillna.html

# Resample data for every 10 min interval to match with LSI meter data format
# Upsample to per minute intervals --> fillna --> interpolate linearly --> downsample to 10 minute intervals
res = data.resample('1T').fillna(method='nearest', limit=1).interpolate('linear').resample('10T').asfreq().dropna()
res
```

```python
# save processed data locally for verification
out_file = "data/processed_node99.csv"
res.to_csv(out_file, index_label = ["timestamp"])
```

```python
df1 = pd.read_csv(out_file)
df1

df2 = pd.read_csv("data/LSI_meter.csv")
df2

plt.plot(df1['t_avg'], 'b')
plt.plot(df2['t_avg'], 'b--')
plt.show()
```

```python
# create a combined df for exporting to csv
combined_df = pd.concat([df2['timestamp'], df2['t_avg'], df1['t_avg']], axis=1, keys=['timestamp', 'LSI_data', 'node99_data'])
combined_df

# save locally
# combined_df.to_csv("data/resampled_node99_and_LSI.csv")
```
